Hadamard gate when applied to a qubit puts a qubit into superposition

```python
from qiskit import *
```

```python
#create a circuit
circuit = QuantumCircuit(1,1)
```

```python
circuit.draw()
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
#call simulator
simulator = Aer.get_backend('statevector_simulator')
```

```python
#execute the curcuit and gather result
result = execute(circuit, backend = simulator).result()
```

```python
#get the statevector from the result
statevector = result.get_statevector()
```

```python
#plot the statevector(circuit without applying any gate)
plot_bloch_multivector(statevector)
```

Now, lets apply the hadamard gate to the qubit to put it into superposition

```python
circuit = QuantumCircuit(1,1)
#apply hadamard gate to the qubit
circuit.h(0)
```

```python
result = execute(circuit, backend = simulator).result()
```

```python
statevector = result.get_statevector()
```

```python
plot_bloch_multivector(statevector)
```

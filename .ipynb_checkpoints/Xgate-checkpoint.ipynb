```python
from qiskit import *
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
#define a quantum circuit with 1 qubit and 1 classical bit
circuit = QuantumCircuit(1,1)
```

```python
#apply X gate on the qubit
circuit.x(0)
```

```python
circuit.draw()
```

```python
#call simulator
simulator = Aer.get_backend('statevector_simulator')
```

```python
#execute circuit on the simulator
result = execute(circuit, backend = simulator).result()
```

```python
#get statevector from result
statevector = result.get_statevector()
print(statevector)
```

```python
#visualize the circuit
```

```python
plot_bloch_multivector(statevector)
```

Adding another X gate on the qubit

```python
circuit.x(0)
```

```python
circuit.draw()
```

```python
result = execute(circuit, backend = simulator).result()
```

```python
statevector = result.get_statevector()
print(statevector)
```

```python
plot_bloch_multivector(statevector)
```

```python
#measure the circuit
circuit.measure([0],[0])    #Measure qubit[0] into classical bit [0], first position -> qubit, second position -> calssical bit
```

```python
circuit.draw()
```

```python
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
```

```python
#plot counts
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

Matrix representation of the circuit:
Use Unitary simulator instead of stateventor simulator

```python
circuit = QuantumCircuit(1,1)
circuit.x(0)
simulator = Aer.get_backend('unitary_simulator')
result = execute(circuit, backend = simulator).result()
unitaryvector = result.get_unitary()
print(unitaryvector)

```
